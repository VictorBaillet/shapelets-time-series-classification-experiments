# Dataset Preprocessing

This Jupyter Notebook is dedicated to the preprocessing of various time series datasets. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
raw_data_path = "raw_datasets"
preprocessed_data_path = "preprocessed_datasets"
```

## Synthetic dataset

```python
np.random.seed(42)
```

```python
N_train = 100 ## Number of time series in the train and test dataset
N_test = 1000
k = 50 ## Length of each time series 
```

```python
motif_1 = np.concatenate((np.zeros(2), 7*np.ones(14), np.zeros(2)))
motif_2 = np.concatenate((np.zeros(2), (np.arange(8)), (6 - np.arange(7)), np.zeros(1),))
```

```python
plt.plot(motif_1, label="Motif 1")
plt.plot(motif_2, label="Motif 2")
plt.legend()
```

```python
x_train = 1.5*np.random.randn(N_train, k)
x_test = 1.5*np.random.randn(N_test, k)
y_train = np.zeros(N_train)
y_test = np.zeros(N_test)
max_motif_length = max(len(motif_1), len(motif_2))

for i in range(N_train):
    idx = np.random.randint(k-max_motif_length-1)
    if i <= N_train//2:
        x_train[i][idx:idx+len(motif_1)] += motif_1
    else:
        y_train[i] = 1
        x_train[i][idx:idx+len(motif_2)] += motif_2
        
for i in range(N_test):
    idx = np.random.randint(k-max_motif_length-1)
    if i <= N_test//2:
        x_test[i][idx:idx+len(motif_1)] += motif_1
    else:
        y_test[i] = 1
        x_test[i][idx:idx+len(motif_2)] += motif_2
```

```python
plt.plot(x_train[65])
```

```python
np.savez_compressed(preprocessed_data_path + '/synthetic_dataset.npz', array1=x_train, array2=y_train, array3=x_test, array4=y_test)
```
